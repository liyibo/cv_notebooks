## 1、房地产政策事件提取

### 主函数文件：main_ht_api.py  

```
def test():
    new = {
        "title": "国务院办公厅转发建设部等部门关于调整住房供应结构稳定住房价格意见的通知",
        "content": "对各地2005年以来落实中央关于房地产市场调控政策的情况进行一次集中检查,少数城市房价上涨过快,房地产市场秩序比较混乱,并根据房地产市场的新情况对部分政策措施作适当调整,（四）严格房地产开发信贷条件,个人住房按揭贷款首付款比例不得低于30%,科学确定房地产开发土地供应规模,继续停止别墅类房地产开发项目土地供应",
    }
    test_worker = Worker(json.dumps({'base_path': '/home/liyibo/demo/nlp-algorithms/config/ht_policy'}))
    
    data_event = test_worker.test(json.dumps(new))

    print(data_event)
```

###  代码初始化参数：

{'base_path': '/home/liyibo/demo/nlp-algorithms/config/ht_policy'}

### process 运行参数：

```
new = {
    "title": "国务院办公厅转发建设部等部门关于调整住房供应结构稳定住房价格意见的通知",
    "content": "对各地2005年以来落实中央关于房地产市场调控政策的情况进行一次集中检查,少数城市房价上涨过快,房地产市场秩序比较混乱,并根据房地产市场的新情况对部分政策措施作适当调整,（四）严格房地产开发信贷条件,个人住房按揭贷款首付款比例不得低于30%,科学确定房地产开发土地供应规模,继续停止别墅类房地产开发项目土地供应",
}
```

### 返回参数：

1、若新闻有事件

"event" 是list，可能有多个事件
```
{
	"message": [{
		"event": ["房地产限购政策"],
		"title": "国务院办公厅转发建设部等部门关于调整住房供应结构稳定住房价格意见的通知"
	}],
	"stat": 0
}
```

2、 新闻无事件

```
{
	"message": [],
	"stat": 0
}
```

3、 若处理有异常

```
{
	"message": [],
	"stat": 1
}
```

## 2、关联新闻和政策事件  

### 主函数文件：policy_connect_api.py  

```
def test():
    new = {
        "title": "中共中央 国务院关于进一步加强城市规划建设管理工作的若干意见",
        "content": "对各地2005年以来落实中央关于房地产市场调控政策的情况进行一次集中检查,少数城市房价上涨过快,房地产市场秩序比较混乱,并根据房地产市场的新情况对部分政策措施作适当调整,（四）严格房地产开发信贷条件,个人住房按揭贷款首付款比例不得低于30%,科学确定房地产开发土地供应规模,继续停止别墅类房地产开发项目土地供应",
    }

    conf_data = {}
    conf_data['base_path'] = "./config/policy_event"
    conf_data['policy_result_table'] = 'policy_result'
    conf_data['news_table'] = 'policy'
    conf_data['db_jar_file'] = "./lib/jar/data-access-service-all-0.0.1.jar"

    test_worker = Worker(json.dumps(conf_data))

    data_event = test_worker.test(json.dumps(new))

    print(data_event)
```

###  代码初始化参数：

'news_table' 里面需要有 "title" 和 "url"字段
```
    conf_data = {}
    conf_data['base_path'] = "./config/policy_event" # 数据路径
    conf_data['policy_result_table'] = 'policy_result' # 房地产政策结果输出表
    conf_data['news_table'] = 'policy' #房地产政策新闻表
    conf_data['db_jar_file'] = "./lib/jar/data-access-service-all-0.0.1.jar"
```
### process 运行参数：


```
    new = {
        "title": "中共中央 国务院关于进一步加强城市规划建设管理工作的若干意见",
        "content": "对各地2005年以来落实中央关于房地产市场调控政策的情况进行一次集中检查,少数城市房价上涨过快,房地产市场秩序比较混乱,并根据房地产市场的新情况对部分政策措施作适当调整,（四）严格房地产开发信贷条件,个人住房按揭贷款首付款比例不得低于30%,科学确定房地产开发土地供应规模,继续停止别墅类房地产开发项目土地供应",
    }
```

### 返回参数：

1、若新闻能关联到政策事件

"message" 中每条对应一个政策事件，[82, "棚改政策"]，82 是政策事件在 policy_result 表中的 id ，"棚改政策" 是该 id 对应的事件名称  

```
{
	"message": [
		[82, "棚改政策"],
		[159, "棚改政策"],
		[730, "棚改政策"]
	],
	"stat": 0
}
```

2、 新闻不能关联到政策事件

```
{
	"message": [],
	"stat": 0
}
```

3、 若处理有异常

```
{
	"message": [],
	"stat": 1
}
```
