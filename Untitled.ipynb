
## 房地产政策事件提取

## Note：

**运行事件提取模块之前需要先部署 "./policy_jar/policyee-1.0.jar"服务，部署方法如下：**

**部署说明:**

**http服务的端口可以通过 application.properties 文件 server.port=8090进行调整。**  

**application.properties 文件位置在 jar 包解压后的目录中："./policyee-1.0/BOOT-INF/classes/application.properties"**

**Linux服务器部署命令：nohup java -jar -Dfile.encoding=utf-8 policyee-1.0.jar &** 

**部署成功后需替换 "./nlp-algorithms/config/ht_policy/event_api.txt" 中的 http 服务地址，当前的服务地址为"http://192.168.2.60:8090/bnp/event"**

### 返回参数：

1、若新闻有事件

"event" 是list，可能有多个事件
```
{
	"message": [{
		"event": ["房地产限购政策"],
		"title": "国务院办公厅转发建设部等部门关于调整住房供应结构稳定住房价格意见的通知"
	}],
	"stat": 0
}
```

2、 新闻无事件

```
{
	"message": [],
	"stat": 0
}
```

3、 若处理有异常

```
{
	"message": [],
	"stat": 1
}
```
