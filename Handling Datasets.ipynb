## 1. Data Augmentation  

在深度学习中，有的时候训练集不够多，或某一类数据较少的时候，可以采用 Data Augmentation 的方法对数据进行扩充，可以有效的防止过拟合，提高模型泛化能力。  

常见的方法：  

Color Jittering：图像亮度、饱和度、对比度变化；

PCA Jittering：首先按照 RGB 三个颜色通道计算均值和标准差，再在整个训练集上计算协方差矩阵，进行特征分解，得到特征向量和特征值，用来做 PCA Jittering；

Random Scale：尺度变换；

Random Crop：采用随机图像差值方式，对图像进行裁剪、缩放；

Horizontal/Vertical Flip：水平/垂直翻转；

Shift：平移变换；

Rotation/Reflection：旋转/仿射变换；

Noise：高斯噪声、模糊处理；

Label shuffle：类别不平衡数据的增广。


下面将介绍如何使用 keras 对数据集进行 data preparation and data augmentation：  

### 1.1 数据介绍

这里用 MNIST 手写数字识别数据集作为实例进行展示，开始先来看一下数据集中的数据。

```python
import numpy as np
from matplotlib import pyplot

def load_data(path="F:/toolbox/cv_notebooks/datas/mnist.npz"):
    f = np.load(path)
    x_train, y_train = f['x_train'], f['y_train']
    x_test, y_test = f['x_test'], f['y_test']
    f.close()
    return (x_train, y_train), (x_test, y_test)

(X_train, y_train), (X_test, y_test) = load_data()
# create a grid of 3x3 images
for i in range(0, 9):
    pyplot.subplot(330 + 1 + i)
    pyplot.imshow(X_train[i], cmap=pyplot.get_cmap('gray'))
# show the plot
pyplot.show()
```

### 1.2 Feature Standardization  

在整个数据集中标准化像素值，即特征标准化，将数据集预处理为 0 均值，单位方差的数据，经过处理后图像数据分布确实变了，但是经过图片显示看起来好像没有变化，不知道什么原因，根据 "Image Augmentation for Deep Learning With Keras" 里面的实例，显示出来的图像跟他博客里面的也不一样，可能是显示参数的问题。。。  

```python
X_train = X_train.reshape(X_train.shape[0], 1, 28, 28)
# convert from int to float
X_train = X_train.astype('float32')
# 去均值
mean = np.mean(X_train, axis=(0, 2, 3))
broadcast_shape = [1, 1, 1]
broadcast_shape[0] = X_train.shape[1]
mean = np.reshape(mean, broadcast_shape)
X_train -= mean

# 除方差
std = np.std(X_train, axis=(0, 2, 3))
broadcast_shape = [1, 1, 1]
broadcast_shape[0] = X_train.shape[1]
std = np.reshape(std, broadcast_shape)
X_train /= (std + 0.000001)

# create a grid of 3x3 images
for i in range(0, 9):
    pyplot.subplot(330 + 1 + i)
    pyplot.imshow(X_train[i].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
# show the plot
pyplot.show()
```

### 1.3 ZCA Whitening  

由于图像中相邻像素之间具有很强的相关性，用于训练输入的数据是冗余的。白化的目的就是降低输入的冗余性；白化后输入具有如下性质：(i)特征之间相关性较低；(ii)所有特征具有相同的方差。

图像中较少的冗余旨在更好地将图像中突出的结构和特征提供给学习算法。

通常，使用 PCA 来进行图像白化。另一种名为 ZCA 的技术具有更好的效果，并且可以保持输入图像的维度，使变换后的图像与原始图像更相似。  

```python
from scipy import linalg

flat_x = np.reshape(X_train, (X_train.shape[0], X_train.shape[1] * X_train.shape[2] * X_train.shape[3]))
sigma = np.dot(flat_x.T, flat_x) / flat_x.shape[0]
u, s, _ = linalg.svd(sigma)
principal_components = np.dot(np.dot(u, np.diag(1. / np.sqrt(s + 0.00001))), u.T)

whitex = np.dot(flat_x, principal_components)
# create a grid of 3x3 images
for i in range(0, 9):
    pyplot.subplot(330 + 1 + i)
    pyplot.imshow(whitex[i].reshape(28, 28), cmap ='gray')
# show the plot
pyplot.show()
```

### 1.4 Random Rotations  

有时样本数据中的图像可能在不同的场景中存在一定程度的旋转。  

可以人为的对输入图像进行随机的旋转，进行数据扩充，提高模型泛化能力。

```python
import scipy.ndimage as ndi

def transform_matrix_offset_center(matrix, x, y):
    o_x = float(x) / 2 + 0.5
    o_y = float(y) / 2 + 0.5
    offset_matrix = np.array([[1, 0, o_x], [0, 1, o_y], [0, 0, 1]])
    reset_matrix = np.array([[1, 0, -o_x], [0, 1, -o_y], [0, 0, 1]])
    transform_matrix = np.dot(np.dot(offset_matrix, matrix), reset_matrix)
    return transform_matrix

def apply_transform(x,
                    transform_matrix,
                    channel_axis=0,
                    fill_mode='nearest',
                    cval=0.):
    """Apply the image transformation specified by a matrix.
    """
    x = np.rollaxis(x, channel_axis, 0)
    final_affine_matrix = transform_matrix[:2, :2]
    final_offset = transform_matrix[:2, 2]
    channel_images = [ndi.interpolation.affine_transform(
        x_channel,
        final_affine_matrix,
        final_offset,
        order=0,
        mode=fill_mode,
        cval=cval) for x_channel in x]
    x = np.stack(channel_images, axis=0)
    x = np.rollaxis(x, 0, channel_axis + 1)
    return x

def tran(x):
    theta = np.pi / 180 * np.random.uniform(-90, 90)

    rotation_matrix = np.array([[np.cos(theta), -np.sin(theta), 0],
                                [np.sin(theta), np.cos(theta), 0],
                                [0, 0, 1]])
    transform_matrix = rotation_matrix

    h, w = 28, 28
    transform_matrix = transform_matrix_offset_center(transform_matrix, h, w)
    x = apply_transform(x, transform_matrix, 0,
                        fill_mode='nearest', cval=0.)
    return x

# create a grid of 3x3 images
for i in range(0, 9):
    pyplot.subplot(330 + 1 + i)
    pyplot.imshow(tran(X_train[i]).reshape(28, 28), cmap ='gray')
# show the plot
pyplot.show()
```

### 1.5 Random Shifts  

图像中目标的位置可能以各种不同的方式偏离图像中心位置，所以可以在训练数据中，以一定方式构造偏离图像中心位置的样本，来提高模型泛化能力。

```python
def tran(x):
    shift = 0.2
    tx = np.random.uniform(0.2, 0.2) * 28
    ty = np.random.uniform(-0.2, 0.2) * 28

    shift_matrix = np.array([[1, 0, tx], [0, 1, ty],[0, 0, 1]])
    transform_matrix = shift_matrix

    h, w = 28, 28
    transform_matrix = transform_matrix_offset_center(transform_matrix, h, w)
    x = apply_transform(x, transform_matrix, 0,fill_mode='nearest', cval=0.)
    return x

# create a grid of 3x3 images
for i in range(0, 9):
    pyplot.subplot(330 + 1 + i)
    pyplot.imshow(tran(X_train[i]).reshape(28, 28), cmap ='gray')
# show the plot
pyplot.show()
```

### 1.6 Random Flips  

训练时候随机翻转训练数据，可以提高模型泛化能力。

```python
def flip_axis(x, axis):
    x = np.asarray(x).swapaxes(axis, 0)
    x = x[::-1, ...]
    x = x.swapaxes(0, axis)
    return x

def tran(x):
    if np.random.random() < 0.5:
        x = flip_axis(x, 2)

    if np.random.random() < 0.5:
        x = flip_axis(x, 1)
    return x

# create a grid of 3x3 images
for i in range(0, 9):
    pyplot.subplot(330 + 1 + i)
    pyplot.imshow(tran(X_train[i]).reshape(28, 28), cmap ='gray')
# show the plot
pyplot.show()
```

### 1.7 Random Crop  

从大的图像上随机切规定大小的图像用于训练 (若 crop 掉的部分太多，可用图像的均值作为 pad，补充为需要的图像大小)

```python
def random_crop(image, crop_size):
    height, width = image.shape[1:]
    dy = crop_size
    dx = crop_size
    if width < dx or height < dy:
        return None
    x = np.random.randint(0, width - dx + 1)
    y = np.random.randint(0, height - dy + 1)
    return image[:, y:(y+dy), x:(x+dx)]

# create a grid of 3x3 images
for i in range(0, 9):
    xx = random_crop(X_train[i], 25)
    pyplot.subplot(330 + 1 + i)
    pyplot.imshow(xx.reshape(25, 25), cmap ='gray')
# show the plot
pyplot.show()
```

### 1.7 Noise  

对输入数据添加随机噪声，视情况改变噪声的强度。

```python
import random

def randomGaussian(img, mean=0.2, sigma=0.3):
    def gaussianNoisy(im, mean=0.2, sigma=0.3):
        for _i in range(len(im)):
            im[_i] += random.gauss(mean, sigma)
        return im

    img.flags.writeable = True  # 将数组改为读写模式
    width, height = img.shape
    img_r = gaussianNoisy(img[:, :].flatten(), mean, sigma)
    img[:, :] = img_r.reshape([width, height])
    return img

# create a grid of 3x3 images
for i in range(0, 9):
    xx = randomGaussian(X_train[i].reshape(28, 28))
    pyplot.subplot(330 + 1 + i)
    pyplot.imshow(xx, cmap ='gray')
# show the plot
pyplot.show()
```

## 2. Imbalanced Datasets

在机器学习任务中，类别数据不均衡是一个典型的存在的问题。

样本不平衡往往会导致模型对样本数较多的分类造成过拟合，即总是将样本分到了样本数较多的分类中。绝大多数常见的机器学习算法对于不平衡数据集都不能很好地工作。

因此从数据和算法两个层面总结了当前处理不均衡数据的几种方法。  

### 2.1 数据层面  

当搜集数据的代价不大时，搜集更多的数据，从而让正负样本的比例平衡是最有效的方法。但是应用场景中数据的比例本身就是不平衡时，这种方法并不能解决数据不平衡问题。

#### 2.1.1 下采样  

下采样方法是针对多数类的样本，当数据量足够时就该使用此方法，减少多数类样本数量。最简单的下采样方法是随机地删掉一些样本。下采样的缺点很明显，就是会丢失样本的一些重要信息，不能够充分利用已有的信息。  

#### 2.1.2 上采样  

最原始的上采样方法是复制稀有类的样本, 但是这样做容易导致过学习, 也没有给正样本增加任何新的信息，对 SVM 算法是无效的，因为 SVM 算法是找支持向量，复制正样本并不能改变数据的支持向量。  

改进的上采样方法则采用加入随机高斯噪声或产生新的合成样本等方法。SMOTE 算法是一个最有名的上采样的改进。SMOTE 是为了解决针对原始上采样方法不能给正样本增加新信息的问题。算法的思想是合成新的少数类样本，合成的策略是对每个少数类样本 a，从它的最近邻中随机选一个样本 b，然后在 a、b 之间的连线上随机选一点作为新合成的少数类样本。

下采样和上采样这两种方法相比而言，都没有绝对的优势，可以将这两种方法结合起来使用。  

使用上采样方法来解决不平衡问题时应适当地应用交叉验证，以确保不会出现过拟合问题。  


采样算法往往很容易实现，并且其运行速度快，并且效果也不错，一些经验法则：  

- 考虑对大类下的样本(超过1万、十万甚至更多)进行欠采样，即删除部分样本；  

- 考虑对小类下的样本(1000 个样本中只有 10 个或更少)进行过采样，即添加部分样本的副本；  

- 考虑尝试随机采样与非随机采样两种采样方法；  

- 考虑对各类别尝试不同的采样比例，比一定是1:1，有时候1:1反而不好，因为与现实情况相差甚远；  

- 考虑同时使用过采样与欠采样。

#### 2.1.3 聚类丰富类  

Sergey Quora 提出了一种优雅的方法，他建议不要依赖随机样本来覆盖训练样本的种类，而是对丰富类别的数据进行聚类。保留聚类中心，然后，基于稀有类和保留的聚类中心数据来训练模型。  

首先，对具有大量样本的丰富类进行聚类操作。假设我们使用的方法是 K-Means 聚类算法。可以选择 K 值为稀有类中的数据样本的个数，并将聚类后的中心点以及相应的聚类中心当做富类样本的代表样例，类标与富类类标一致。  

经过上述步骤的聚类操作，对富类训练样本进行了筛选，接下来我们就可以将相等样本数的 K 个正负样本进行有监督训练。  

### 2.2 算法层面  


强烈建议不要对待每一个分类都使用自己喜欢而熟悉的分类算法。应该使用不同的算法对其进行比较，因为不同的算法使用于不同的任务与数据。具体可以参见“Why you should be Spot-Checking Algorithms on your Machine Learning Problems”。 
  决策树往往在类别不均衡数据上表现不错。它使用基于类变量的划分规则去创建分类树，因此可以强制地将不同类别的样本分开。目前流行的决策树算法有：C4.5、C5.0、CART和Random Forest等。基于R编写的决策树参见这里。基于Python的Scikit-learn的CART使用参见这里。

#### 2.2.1 转化为一分类问题  

对于二分类问题，如果正负样本分布比例极不平衡，可以换一个完全不同的角度来看待问题：把它看做一分类或异常检测问题。这类方法的重点不在于捕捉类间的差别，而是为其中一类进行建模，经典的工作包括 One-class SVM等。One Class SVM 是指训练数据只有一类正样本的数据，而没有另外的一类。在这时，需要学习的实际上训练数据的边界。对于正负样本极不均匀的问题，使用异常检测，或者一分类问题，也是一个思路。  

#### 2.2.2 分类器集成方法  

- EasyEnsemble：  

训练多个模型，每个模型使用稀有类别的所有样本和丰富类别的 n 个不同样本。假设存在 1000 例稀有类别和 10000 例丰富类别。将10000 例丰富类别随机采样为 10 份数据，这样与 1000 例稀有类别数据合并起来，就可以训练 10 个模型。  


- BalanceCascad：  

这个方法跟 EasyEnsemble 有点像，但不同的是，每次训练 adaboost 后都会扔掉已被正确分类的样本，经过不断地扔掉样本后，数据就会逐渐平衡。 

#### 2.2.3 改变样本权重  

在大部分不平衡分类问题中，稀有类是分类的重点。在这种情况下，正确识别出稀有类的样本比识别大类的样本更有价值。反过来说，错分稀有类的样本需要付出更大的代价。  

改变样本权重指的是增大样本数较少类别的样本的权重，当这样的样本被误分时，其损失值要乘上相应的权重，从而让分类器更加关注这一类数目较少的样本。  

#### 2.2.4 特征选择方法  

特征选择方法对于不平衡分类问题同样具有重要意义。样本数量分布很不平衡时，特征的分布同样会不平衡。尤其在文本分类问题中，在大类中经常出现的特征，也许在稀有类中根本不出现。因此，根据不平衡分类问题的特点，选取最具有区分能力的特征，有利于提高稀有类的识别率。  

### 2.3 评价指标  

在评估分类器的性能和指导分类器建模方面，评估标准发挥了关键作用。在传统的分类方法中，准确率是常用的指标。然而在不平衡数据分类中，准确率不再是恰当的指标。假如将所有的样本都归为预测为负样本，准确率就会很高，但这样的模型显然是不好的，已经对不平衡的样本过拟合了。  

- 混淆矩阵(Confusion Matrix)：使用一个表格对分类器所预测的类别与其真实的类别的样本统计，分别为：TP、FN、FP与TN。
- 精确度(Precision)
- 召回率(Recall)
- F1得分(F1 Score)：精确度与召回率的加权平均。
- Kappa
- ROC曲线

## 3. Noisy / Unlabelled Data  

### 3.1 Noisy Data  

在监督学习中，训练数据所对应的标签质量对于学习效果至关重要。如果学习时使用的标签数据都是错误的，那么不可能训练出有效的预测模型。同时，深度学习使用的神经网络往往结构复杂，为了得到良好的学习效果，对于带标签的训练数据的数量也有较高要求。

矛盾在于：给数据打标签这个工作在很多场景下需要人工实现，大量、高质量标签本身费时费力，在经济上相对昂贵。因此，实际应用中的机器学习问题必须面对噪音标签的影响，即我们拿到的每一个带标签数据集都要假定其中是包含噪声的。进一步，由于样本量很大，对于每一个带标签数据集，我们不可能人工逐个检查并校正标签。

基于上述矛盾现状，在实际工作中必须面对以下两点问题

1.	训练集带标签样本中噪音达到什么水平对于模型预测结果会有致命影响

2.	对于任意给定带标签训练集，如何快速找出可能是噪音的样本  

[这篇知乎文章](https://zhuanlan.zhihu.com/p/28703943) 用 MNIST 数据集做了实验，发现在在噪声标签占比超过70%之后，准确率才开始快速的下降，并解释了为什么在噪音标签占比70%的情况下，模型抗噪声性能还比较好的原因。  

原因：在加噪声的第一步，均匀的随机抽取出一定比例的标签，考虑到原始数据 10 类标签的分布是基本均匀的，那么抽出来的 K 个样本中 10 类标签的数量基本一致。在第二步，对于每个标签，将正确标签抹去，从正确标签之外的 9 个字符中选择一个作为标签，由于选择算法本身也是随机的，那么，错误标签是均匀分布在其他 9 类的。这样，正确标签相比其他任何一个类别，仍然占有明显数量优势，所以 CNN 才可以根据这个数量优势学习到正确标签。而当噪声比例进一步增加后，数量对比优势会逐渐弱化，在模型训练中，正确标签带来的梯度下降增益不足以对抗错误标签的影响，神经网络倾向于学习到随机标签，就很难训练出有意义的模型。  

所以在多类别分类任务中，存在一定比例的噪声数据是可以容忍的。但这篇文章实验中噪声数据是在其他类别中均匀分布的，一定程度上增加了模型抗噪声的能力。  

在实际场景中，噪声数据可能是集中在个别标签中的，这个时候模型抗噪声能力就会减弱，有条件的话还需要找出这部分噪声数据，进行重新标注。  

该实验中，通过观察网络中经过 softmax 后的最后一层输出概率值，来快速找出噪声数据，让网络把自己认为存在二义性的样本和标签筛选出来，实现快速找出疑似噪音标签。eg. 分类器认为最有可能类别的概率低于70%，同时第二可能类别概率高于15%, 就认为可能是噪声数据。  

### 3.2 Unlabelled Data  

深度学习任务中，需要面临的一个问题是缺乏大量的标注数据，因此如何使用尽可能少的标签数据来训练一个效果较好的分类器是一个亟待解决的问题。  

**Active Learning**  

有的时候，有类标的数据比较稀少而没有类标的数据是相当丰富的，但是对数据进行人工标注又非常昂贵，这时候，主动学习就可以主动地提出一些标注请求，将一些经过筛选的数据提交给专家进行标注。

主动学习目是通过尽量少的标签，提高准确率，减少获得标签数据的费用。  

主动学习就是去学习那些比较“难的”，“信息量大的”样本。关键点是每次都挑当前分类器分类效果不理想的那些样本 (hard sample) 给它训练，假设是训练这部分 hard sample 对于提升分类器效果最有效而快速。问题是在不知道真正标签的情况下怎么去定义 HARD sample 或者说怎么去描述当前分类器对于不同样本的分类结果的好坏。

- 定义：由于深度学习的输出是属于某一类的概率（0～1），一个很直观的方法就是用“熵（entropy）”来刻画信息量，把那些预测值模棱两可的样本挑出来，对于二分类问题，就是预测值越靠近0.5，它们的信息量越大。还有一个比较直观的方法是用“多样性（diversity）”来刻画 labeled data 和unlabeled data 的相似性。这是两个十分重要的 Active Learning 选择指标。  

具体过程：  

1. 把所有数据送入预训练的网络中测试一遍，得到预测值，然后挑出 hard sample，对这些数据进行标注
2. 用这些刚刚给了标签的样本训练深度网络
3. 把剩下的没有标注的数据送入上面训练过的网络中，得到预测值，还是挑出hard sample，对这些数据进行标注
4. 刚刚给了标签的样本和原来有标签的样本组成新的训练集，继续用它们来训练深度网络
5. 重复步骤3，4，直到循环中断  

Fine-tuning Convolutional Neural Networks for Biomedical Image Analysis: Actively and Incrementally 改进了 Active Learning，提供了一个新型的方法 AIFT（Active，Incremental Fine-Tuning），把主动学习和迁移学习集成到一个框架。AIFT 算法开始是直接使用一个预训练从未标注数据里找一些比较值得标注的样本，然后模型持续的加入新标注的数据，一直做微调。这种方法显著的减低标注的工作量，并且有指导的选择哪些数据需要标注，同时降低了数据增强带来的噪声影响。
